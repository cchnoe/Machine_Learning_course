#### Eligiendo la mejor cantidad de clusters para nuestra data! 

- El gráfico del codo (elbow graph) nos dirá cuánto cambia la varianza interna dentro de mi clusters con respecto al número total de clusters que hay dentro de mi data. Se elige el punto del codo porque es el punto en el que la varianza empieza a desacelerar más o hasta aumentar.  

```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import cdist
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
iris = pd.read_csv('https://raw.githubusercontent.com/HackSpacePeru/Datasets_intro_Data_Science/master/Iris_people/iris.csv')
```

```python
K = range(1,15)
distorciones= []
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(iris.iloc[:,:4])
    distorciones.append(sum(np.min(cdist(iris.iloc[:,:4], kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / iris.iloc[:,:4].shape[0])

# Graficando el codo
plt.plot(K, distorciones, 'bx-')
plt.xlabel('k')
plt.ylabel('Distorciones de la varianza')
plt.title('El método del codo enseñando el valor óptimo de K')
plt.show()
```
